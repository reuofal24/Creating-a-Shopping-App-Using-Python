```python
#Database
admin_DB={"admin1": "123", "admin2": "456"}
customer_DB={"customer1": "987", "customer2": "654"}
products_DB = [
    {"id": 11, "name": "Boots", "category": 1, "price": 200},
    {"id": 22, "name": "Coats", "category": 2, "price": 400},
    {"id": 33, "name": "Jackets", "category": 3, "price": 300},
    {"id": 44, "name": "Caps", "category": 4, "price": 100}
]
category =[
    {"id": 1, "name": "Clothes"},
    {"id": 2, "name": "Shoes"},
    {"id": 3, "name": "Accessories"}
]

str(category)
str(products_DB)

shopping_cart = [] 
shopping_quant= []
```

```python
#inhertinas for users : customer + admin
class user:
    def __inti__(self, username,password,session_id):
        self.username = username
        self.password = password
        self.session_id = session_id
    
        
    def login(self, username, password):
        username = input("Enter the username : ")
        password = input("Enter the password : ")
        if username in admin_DB and admin_DB[username] == password:     
            print("You are admin")
            session_id = 0
            return session_id
        elif username in customer_DB and customer_DB[username] == password:
            print("you are customer")
            session_id = 1
            return session_id
        else:
            print("Invalid user type. Please try again")
        
    def logout(self):
        logut = input("If you want to logout from account enter yes  ")
        if logut == 'yes': 
            print("You have successfully exited")
            exit()
        else:
            print("Invalid chosse")
            
            
class admin(user):
    def __inti__(self, username,password,session_id):
        user.__init__(self,username,password,session_id)
        
    def login(self, username, password):
        self.session_id= super().login(username, password)
    
    def adminAddProduct(self):
        if self.session_id == 0:
            nid = int(input("Enter Product ID: "))
            nname = input("Enter Product Name: ")
            ncategory = int(input("Enter Category ID: "))
            nprice = float(input("Enter Price: "))
            p = [{"id": nid, "name": nname, "category": ncategory, "price": nprice}]
            products_DB.append(p)
            print("Product Added Successfully.")
            print(products_DB)
                          
    def modifyProduct(self):
        if self.session_id == 0:
            op= input("Enter what you want to modify: ")
            up= input("Enter the new value: ")
            for x in products_DB:
                for y in x:
                    if x.get(y) == op:
                        x.update({y: up})
                        print(products_DB)
        
    def adminRemoveProduct(self):
        if self.session_id == 0:
            pid = int(input("Enter Product ID: "))
            for p in range(len(products_DB)):
                if products_DB[p]["id"] == pid:
                    del products_DB[p]
                    print("Product Removed Successfully.")
                    print(products_DB)
                    break
                else:
                    print("Wronge Id Product")
            
    def addCategory(self):
        if self.session_id == 0:
            cid = int(input("Enter Category ID: "))
            cname = input("Enter Category Name: ")
            c = [{"id": cid, "name": cname}]
            category.append(c)
            print("Category Added Successfully.")
            print(category)
        
    def removeCategory(self):
        if self.session_id == 0:
            cid = int(input("Enter Category ID: "))
            for c in range(len(category)):
                if category[c]["id"] == cid:
                    del category[c]
                    print("Category Removed Successfully.")
                    print(category)
                    break
                else:
                    print("Wronge Id Category")

    def adminChoice(self):
        if self.session_id == 0:
            choice = int(input("Please enter a choice: "))
            if choice == 1:
                self.adminAddProduct()
                print("You are added product finshied")
                
            elif choice == 2:
                self.modifyProduct()
                print("You are modify product finshied")

            elif choice == 3:
                self.adminRemoveProduct()
                print("You are removed product finshied")

            elif choice == 4:
                self.addCategory()
                print("You are added category finshied")

            elif choice == 5:
                self.removeCategory()
                print("You are removed category finshied")
                
            elif choice == 6:
                print("Thank you for shopping with us.")
        else:
            print("\nInvalid Choice. Please enter valid choice")
            print("_________")
                 
    def adminMenu(self):
        if self.session_id == 0: 
            print("_________")
            print("1.Add Product")
            print("2.Modify a Product")
            print("3.Remove Product")
            print("4.Add a new Category")
            print("5.Delete a Category")
            print("6.Exit")
            print("_________")
            self.adminChoice() 
    
           
class customer(user):
    def __inti__(self, username,password,session_id):
        user.__init__(self,username,password,session_id)
        
    def login(self, username, password):
        self.session_id= super().login(username, password)
    
    def add_item(self):
        if self.session_id == 1:
            for kk in range(len(products_DB)):
                print(kk+1,products_DB[kk], sep='\t')
            order=int(input("chose a product to add in cart:"))
            if order <= 5:
                # User wants to purchase something.
                print('You selected', products_DB[order-1])
                quant = int(input('How many units do you wish to purchase?\n'))
                if products_DB[order-1] in shopping_cart:
                    idx = shopping_cart.index(products_DB[order-1])
                    shopping_quant[idx]+=quant
                else:
                    shopping_cart.append(products_DB[order-1])        
                    shopping_quant.append(quant)        
                print('Added to shopping cart:', quant, 'units of', products_DB[order-1])
            else: 
                print("Sorry that was not a valid input.")
                
                
    def remove_item(self):
        if self.session_id == 1:
            for kk in range(len(shopping_cart)):
                print(f"{shopping_cart[kk]['id']}: {shopping_cart[kk]['name']} ({shopping_quant[kk]})")

            cid = int(input("Enter product id: "))

            # Find the index of the product with the given product ID
            index_to_remove = -1
            for i in range(len(shopping_cart)):
                if shopping_cart[i]['id'] == cid:
                    index_to_remove = i
                    break

            if index_to_remove != -1:
                quantity_to_remove = int(input("Enter the quantity to remove: "))
                if shopping_quant[index_to_remove] >= quantity_to_remove:
                    shopping_quant[index_to_remove] -= quantity_to_remove
                    print("Product removed successfully.")

                    # Check if the quantity of the product is now zero, and remove it from the cart if so
                    if shopping_quant[index_to_remove] == 0:
                        del shopping_cart[index_to_remove]
                        del shopping_quant[index_to_remove]
                        print("Product removed from the cart as it has no quantity left.")
                else:
                    print("The quantity you entered is more than what is in the cart.")
            else:
                print("Product not found in the cart.")

    
    def viewCart(self):   ##not existed
        if self.session_id == 1:
            print('Your Shopping Cart:')
            for kk in range(len(shopping_cart)):
                print(shopping_cart[kk],"quantity", shopping_quant[kk])
    
    def checkout(self):
        if self.session_id == 1:
            tot_price=0.0
            for product_name in shopping_cart:
            # Check if the product name exists in the list of products
                if product_name in products_DB:
                    for kk in range(len(shopping_cart)):
                        idx = products_DB.index(shopping_cart[kk])
                        unit_price=products_DB[idx]["price"]
                        tot_price = round(unit_price*shopping_quant[kk], 2)
                        #grand_tot += tot_price

            print("your total is: ",tot_price)
            print()
            print("1: Net Banking")
            print("2: Credit Card")
            print("3: UPI")
            option = int(input("Enter an Option to make the payment:"))

            if option == 1:
                print("Your order is successfully placed with Net Banking..")
            elif option == 2:
                print("Your order is successfully placed with Credit Card...")
            elif option == 3:
                print("Your order is successfully placed with UPI")
            else:
                print("An Invalid Payment Option")
    
    def customerChoice(self):
        if self.session_id == 1:
            choice = int(input("Please enter a choice: "))
            if choice == 1:
                self.add_item()
                print("You are added product finshied")
                
            elif choice == 2:
                self.remove_item()                
            elif choice == 3:
                self.viewCart()
                print("You are viewed cart finshied")
    
            elif choice == 4:
                self.checkout()
                print("You are checked finshied")
                
            elif choice == 5:
                print("Thank you for shopping with us.")
        else:
            print("\nInvalid Choice. Please enter valid choice")
            print("_________")
            
    def customerMenu(self):
        if self.session_id == 1:
            print("_________")
            print("1.Add Product")
            print("2.Remove Product")
            print("3.View Cart")
            print("4.Check out")
            print("5.Exit")
            print("_________")
            self.customerChoice()
    
print("Welcome to the Demo Marketplace")

x=admin()
x.login("admin1","123")
if x.session_id == 0 :
    x.adminMenu()
else:
    print(' ')    
#x.logout()

x=customer()
x.login("customer1","987")
if x.session_id == 1 :
    x.customerMenu()
else:
    print(' ')
#x.logout()
```
